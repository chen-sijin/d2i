```python
n = 100
trace = []
rel_err = []
for i in range(n):
    np.random.seed(i)
    mat_test = np.random.multivariate_normal(mean = np.zeros(cov.shape[0]), cov = cov, size = 50)
    C_test = np.cov(mat_test, rowvar=False)
    trace.append(np.trace(prc_mat @ C_test))

trace = np.array(trace)
    
```

```python
a_list = []
a_inv_list = []
n = 1000
for i in range(n):
    np.random.seed(i)
    ran = np.random.uniform(0, 1, 10000)
    a_list.append(np.cov(ran)/(1/12))
    a_inv_list.append(1/12/np.cov(1/ran))
```

```python
plt.plot(np.arange(n), a_list)
plt.plot(np.arange(n), np.mean(a_list)*np.ones_like(a_list), label = f"{np.mean(a_list):.3e}")
plt.legend()
```

```python
plt.plot(np.arange(n), a_inv_list)
plt.plot(np.arange(n), np.mean(a_inv_list)*np.ones_like(a_list), label = f"{np.mean(a_inv_list):.3e}")
plt.legend()
```

# Problem 2

```python
alpha_true = 0
sigma = 0.5
n_samples = 1000

# N = 1001

# d_hat=[]
# for i in range(N):
#     np.random.seed(i)
#     rv = np.random.normal(0, sigma, n_samples)
#     d_hat.append(np.mean(rv))

np.random.seed(10086)
d_hat = np.random.normal(0, sigma, n_samples)


```

```python
# print(d_hat)
```

```python
plt.hist(d_hat, bins = np.linspace(-2, 2, 101), edgecolor="black")
```

```python
N = 1001

alpha_range = np.linspace(-1, 1, N)
```

```python
posterior = 1 / (np.sqrt(2 * np.pi * sigma)) * np.exp(-0.5 * (np.mean(d_hat) - alpha_range**3)/sigma*(np.mean(d_hat) - alpha_range**3))
```

```python
print(posterior.shape)
```

```python
plt.hist(posterior, bins = np.linspace(-1, 1, 50), edgecolor="black")
plt.xlabel('alpha')
```

```python
plt.plot(np.arange(len(posterior)), posterior)
```

```python
# Now the normal way of giving data is passing a numpy array and parameter separately
c = ChainConsumer().add_chain(posterior, parameters=['alpha']).configure(statistics="max")
fig = c.plotter.plot()
fig.set_size_inches(3 + fig.get_size_inches())  # Resize fig for doco. You don't need this.
```

```python
import numpy as np
import matplotlib.pyplot as plt


```

```python
a_true = 3
v = np.array([2, 5])
cov = np.array([[1, -1], [-1, 5]])
```

```python
print(a_true*v)
```

```python
# np.random.seed(114514)

rng = np.random.default_rng(114514)

x =rng.multivariate_normal(a_true * v, cov, 10000)
```

```python
print(x[:10])
```

```python
print(np.mean(x[:,0]))
print(np.cov(x[:,0]))
print(np.mean(x[:,1]))
print(np.cov(x[:,1]))
```

```python
plt.scatter(x[:,0], x[:, 1], alpha=0.3, s=20, color="#4292c6")
plt.grid()
plt.scatter(6, 15, color = "#084594", s=60)
# plt.axis('scaled')
```

$$\mathrm{Cov}\left( A_{best} \right) =\boldsymbol{V}^{-1}\boldsymbol{C}\left( \boldsymbol{V}^T \right) ^{-1}$$

```python
a  = np.zeros(len(x))
for i in range(len(x)):
    a[i] = x[i].T @ np.linalg.inv(cov) @ v / (v.T @ np.linalg.inv(cov) @ v)
```

```python
plt.figure(figsize = (20,4))

plt.plot(np.arange(len(a)), a, alpha=0.4, color = 'orange', label = "$a_{best}$")
plt.hlines(xmin=0, xmax = len(x)-1, y=a_true, label = "$a_{true}$")
plt.xlim((0, len(x)))
plt.legend()
```

```python
print(np.cov(a))
```

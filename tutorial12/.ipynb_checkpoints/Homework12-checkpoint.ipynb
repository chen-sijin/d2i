```python
import numpy as np
import matplotlib.pyplot as plt 
from chainconsumer import ChainConsumer
from tqdm import trange 
```

For a multi-varaite Gaussian distribution:
$$x \sim\mathcal{N} \left( \mu ,\Sigma \right) $$

then, for a conditional multi-variate gaussian distribution:
$$x_1|x_2\sim\mathcal{N} \left( \mu _{1|2}, \Sigma _{1|2} \right) $$
where, we have $\mu =\left[ \begin{array}{c}
	\mu _1\\
	\mu _2\\
\end{array} \right] $, $\Sigma =\left[ \begin{matrix}
	\Sigma _{11}&		\Sigma _{12}\\
	\Sigma _{21}&		\Sigma _{22}\\
\end{matrix} \right] $. 

The mean value and covariance matrix of the conditional PDF is :
$$\mu _{1|2}=\mu _1+\Sigma _{12}\Sigma _{22}^{-1}\left( x_2-\mu _2 \right) $$
$$\Sigma _{1|2}=\Sigma _{11}-\Sigma _{12}\Sigma _{22}^{-1}\Sigma _{21}$$

For the joint distribution, we have:
\begin{equation}
p\left( \mu ,t,x \right) =p\left( \mu \right) \cdot p\left( t|\mu \right) \cdot p\left( x|t \right) \propto e^{-\frac{1}{2}\left( t-\mu \right) ^2}e^{-\frac{1}{2}\left( x-t \right) ^2}
\end{equation}

For the conditional PDF of $\mu$, we have:
\begin{align}
    \begin{split}
        p\left( \mu|t ,x \right) &=\frac{p\left( t,\mu ,x \right)}{p\left( \mu ,x \right)}
\\
&=\frac{p\left( t,\mu ,x \right)}{\int{p\left( t,\mu ,x \right) d\mu}}
    \end{split}
\end{align}

So, in the denominator, we could calculate:
\begin{align}
    \begin{split}
        \int{p\left( t,\mu ,x \right) d\mu}=\int{e^{-\frac{1}{2}\left( t-\mu \right) ^2}e^{-\frac{1}{2}\left( x-t \right) ^2}d\mu}\propto e^{-\frac{1}{2}\left( x-t \right) ^2}
    \end{split}
\end{align}

So, the conditional PDF of $\mu$ becomes:
\begin{align}
    \begin{split}
        p\left( \mu |t,x \right) &=\frac{p\left( t,\mu ,x \right)}{p\left( t ,x \right)}
\\
&=\frac{p\left( t,\mu ,x \right)}{\int{p\left( t,\mu ,x \right) d\mu}}
\\
&\propto \frac{e^{-\frac{1}{2}\left( t-\mu \right) ^2}e^{-\frac{1}{2}\left( x-t \right) ^2}}{e^{-\frac{1}{2}\left( x-t \right) ^2}}
\\
&\propto e^{-\frac{1}{2}\left( t-\mu \right) ^2}
    \end{split}
\end{align}


Then, for the conditional PDF of $t$, we have:
\begin{align}
    \begin{split}
        p\left( t|\mu ,x \right) &=\frac{p\left( t,\mu ,x \right)}{p\left( \mu ,x \right)}
\\
&=\frac{p\left( t,\mu ,x \right)}{\int{p\left( t,\mu ,x \right) dt}}
    \end{split}
\end{align}

The integral of the denominator is :
\begin{align}
    \begin{split}
        \int{p\left( t,\mu ,x \right) dt}&=\int{e^{-\frac{1}{2}\left( t-\mu \right) ^2}e^{-\frac{1}{2}\left( x-t \right) ^2}dt}
\\
&=\int{e^{-\frac{1}{2}\left( t^2-2t\mu +\mu ^2+x^2-2tx+t^2 \right)}dt}
\\
&=\int{e^{-\frac{1}{2}\cdot 2\left[ t^2-\left( \mu +x \right) t+\left( \mu +x \right) ^2/4 \right]}e^{-\frac{1}{2}\left( x^2+\mu ^2 \right)}e^{\frac{\left( x+\mu \right) ^2}{4}}dt}
\\
&\propto e^{-\frac{1}{2}\left( x^2+\mu ^2 \right)}e^{\frac{\left( x+\mu \right) ^2}{4}}
    \end{split}
\end{align}

Therefore, the original conditional PDF becomes:
\begin{align}
    \begin{split}
        p\left( t|\mu ,x \right) &=\frac{p\left( t,\mu ,x \right)}{p\left( \mu ,x \right)}
\\
&=\frac{p\left( t,\mu ,x \right)}{\int{p\left( t,\mu ,x \right) dt}}
\\
&\propto \frac{e^{-\frac{1}{2}\left( t-\mu \right) ^2}e^{-\frac{1}{2}\left( x-t \right) ^2}}{e^{-\frac{1}{2}\left( x^2+\mu ^2 \right)}e^{\frac{\left( x+\mu \right) ^2}{4}}}
\\
&=e^{-\frac{1}{2}\cdot 2\left[ t-\left( \frac{\mu +x}{2} \right) \right] ^2}
    \end{split}
\end{align}

```python
nsamples = 1000000

mu = np.zeros(nsamples)
t = np.zeros(nsamples)

x_obs = 2

for i in trange(1, nsamples):
    t[i] = np.random.normal((mu[i-1]+x_obs)/2, 1/np.sqrt(2))
    mu[i] = np.random.normal(t[i], 1)

    
    

```

```python
plt.figure()
plt.hist(t, bins=500, density=True);
plt.xlabel("$t$")


```

```python
plt.figure()
plt.hist(mu, bins=500, density=True);
plt.xlabel("$\mu$")
```

```python
joint_t_mu = np.array([mu, t])
joint_t_mu = joint_t_mu.T
print(joint_t_mu.shape)


c = ChainConsumer().add_chain(joint_t_mu[500000:, :], parameters=["$\mu$", "$t$"])
fig = c.plotter.plot()
fig.set_size_inches(3 + fig.get_size_inches())  # Resize fig for doco. You don't need this.
```
